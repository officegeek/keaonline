# Import scipy and draw the line of Linear Regression

```python
import matplotlib.pyplot as plt
from scipy import stats
```

## Method that returns key values of Linear Regression:

- slope
- intercept
- r
- p
- std_err

```python
slope, intercept, r, p, std_err = stats.linregress(x, y)
```

### Create a function (*myfunc*) that uses the slope and intercept values to return a new value. This new value represents where on the y-axis the corresponding x value will be placed:

```python
def myfunc(x):
  return slope * x + intercept
```

### Run each value of the x array through the function. This will result in a new array with new values for the y-axis:

```python
mymodel = list(map(myfunc, x))
```

```python
plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
```
